## Load and visualize and image

```python
%%capture
#@markdown <a href="https://abraia.me/console/gallery" target="_blank">Upload and manage your images</a>

url = 'http://upload.wikimedia.org/wikipedia/commons/1/13/Usain_Bolt_16082009_Berlin.JPG'
abraia.upload(url, 'usain.jpg')

f = abraia.load('usain.jpg')
meta = abraia.metadata('usain.jpg')

print(meta)
```

```python
from PIL import Image

import numpy as np
import matplotlib.pyplot as plt

img = np.asarray(Image.open(f))

plt.figure()
plt.title('Image')
plt.imshow(img)
plt.axis('off')
plt.show()
```
