```python
%%capture
#@markdown <a href="https://abraia.me/console/gallery" target="_blank">Upload and manage your images</a>

url = 'http://upload.wikimedia.org/wikipedia/commons/1/13/Usain_Bolt_16082009_Berlin.JPG'
abraia.upload_file(url, 'usain.jpg')
```

## List files

List the files and folders in the root folder (`''`).

```python
import pandas as pd

folder = ''
files, folders = abraia.list(folder)
pd.DataFrame(files)
```

## Plot an image

Load an image as numpy array and show it in a plot.

```python
import matplotlib.pyplot as plt

img = abraia.load_image('usain.jpg')

plt.figure()
plt.title('Image')
plt.imshow(img)
plt.axis('off')
plt.show()
```

## Image metadata

Read the image metadata and save it as a JSON file.

```python
import json

meta = abraia.load_metadata('usain.jpg')
abraia.save_file('usain.json', json.dumps(meta))
meta
```
